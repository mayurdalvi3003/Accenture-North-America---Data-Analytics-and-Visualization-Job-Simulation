```python
plt.bar(x , y )
plt.title("Top-5 Categories based on Popularity score")
plt.xlabel("Category ----->")
plt.ylabel("Popularity Score ------>")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = new_df1.groupby("Category")["Score"].sum().sort_values(ascending = False).head().index
y = new_df1.groupby("Category")["Score"].sum().sort_values(ascending = False).head().values

plt.pie(y, labels=x, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Popularity percentage of top-5 categories')

plt.show()
```

```python
#  How many reactions are there to the most popular category?	
x = new_df1.groupby("Category")["Reaction Type"].count().sort_values(ascending = False).head().index
y = new_df1.groupby("Category")["Reaction Type"].count().sort_values(ascending = False).head().values

plt.bar(x , y )
plt.title("Count of Reaction for the Top-5 categories ")
plt.xlabel("Category ----->")
plt.ylabel("Reaction Counts ------>")
plt.show()
```
